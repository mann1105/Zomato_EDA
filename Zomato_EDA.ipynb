```python
df_country = pd.read_csv("Country-Code.csv")
print(df_country.head())
```

```python
final_df = pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head()
```

```python
final_df.head(2)
```

```python
final_df.dtypes
```

```python
country_names = final_df['Country'].value_counts().index
print(country_names)

#first country represents most users
```

```python
country_val = final_df['Country'].value_counts().values
```

```python
plt.pie(country_val[:3],labels=country_names[:3],autopct='%.2f%%')
```

# Observation

Zomato maximum records/transactions are from India, US and UK

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

# Observations

1. When rating btw 4.5 to 5 food was excellant

2. 4.0 to 4.4 --> very good
3. 3.5 to 3.0 --> good
4. 3.0 to 3.4 --> Average
5. 2.5 to 2.9 --> Average

```python
ratings.head()
```

```python
plt.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings,hue = 'Rating color',palette=['white','red','orange','yellow','green','green'])
```

# Observation

1. not rated very high
2. Maximum number of ratign are btw 2.5 to 3.5

if any values are missing then use btw above

```python
sns.countplot(x = "Rating color",palette=['white','red','orange','yellow','green','green'],data = ratings)
```
